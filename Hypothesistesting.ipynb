1.  What is a null hypothesis (H₀) and why is it important in hypothesis testing?

Answer - The null hypothesis is a fundamental assumpation in statistics that states there is no significant difference, effect or relationship between the variables being studied.

Importance of null hypothesis -

* Provides a baseline for testing
* Helps in determine if results aree due to chance
* Framework for statistical decision-making



2. What does the significance level (α) represent in hypothesis testing?

Answer - In hypothesis testing, the significance level (a) represents the probability of committing a type 1 error. This occurs when we mistakenly reject a null hypothesis that is actually true. Commonly useed levels are 0.01, 0.05 and 0.10 depending on how much certainty the speacific field of study requires.

3. Differentiate between Type I and Type II errors.

Answer - Hypothesis testing errors categorized into two types -

Type 1 error - It occurs when the null hypothesis is true, but we reject it. This is often called a "false positive" - wee claim there is an effect or a difference when, in reality there is not one.

It is denoted by alpha which is the significance level of the best.

Type 2 error - It occurs when the null hypothesis is falsebut we fail to reject it. This is known as a "false negative" - we miss an effect or a difference that actually exists.

It is denoted by beta. The power of test is defined as 1-b.


4.  Explain the difference between a one-tailed and two-tailed test. Give an example of each.

Answer - Difference between one-tailed and two tailed test -

One tailed test - It is used when we are interested in a change in only one direction. we are testing if the sample statistic is significantly higher or significantly lower than the population parameter, but not both.

Example - A company claims that sales increases this year.

Two tailed test - it is used when we want to determine if there is any significant difference, regardless of the direction. We are testing if the sample statistic is simply different from the population parameter.

Example - A new teaching method changes CBSE exam scores. Old scores is not equal current scores.
